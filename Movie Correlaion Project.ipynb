```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#loading the data set
df = pd.read_csv(r'C:\Users\eltanany shop\Desktop\IEEE\movies.csv')
```

```python
#showing the first five rows
df.head()
```

```python
#showing the sum of the null values
df.isnull().sum()
```

```python
#showing the data type of every column 
df.dtypes
```

```python
#delete null values
df = df.dropna()
df.isnull().sum()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#to replace budget,gross from float to int
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.dtypes
```

```python
# create correct year column in new column
df['yearcorrect'] = df['released'].astype(str).str[0:4]
df
```

```python
#sorting the values of column of gross from the large value to the small value
df = df.sort_values(by = ['gross'], inplace = False , ascending= False)
df
```

```python
# drop any duplicates
df.drop_duplicates()
```

```python
# Budget vs Gross Earning
plt.scatter(x = df['budget'], y = df['gross'],marker='.',color = 'g')

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earning')
plt.ylabel('Budget')
plt.show()
```

```python
# plot Budget vs Gross using seaborn

sns.regplot( x = 'budget' , y = 'gross', data = df,scatter_kws = {'color':'red'},line_kws = {'color':'blue'} ,marker='.')
```

```python
#lets look starting to show correlation
df.corr(method = 'pearson')  #method : {'pearson', 'kendall', 'spearman'}
```

```python
#lets look starting to show correlation
df.corr(method = 'pearson')  #method : {'pearson', 'kendall', 'spearman'}
```

```python
df.head()
```

```python
dff = df
```

```python
df_numerized = dff
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
    
```

```python
corr_matrix = df_numerized.corr(method = 'pearson')
plt.figure(figsize = (15,15))
sns.heatmap(corr_matrix , annot = True )

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

correlation_pairs = correlation_mat.unstack()

correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs
```

```python
#showin high correlation
high_corre = sorted_pairs[(sorted_pairs) > 0.5]
high_corre
```
